### Load The Data From The File 

```python
import numpy as np
Data = np.genfromtxt('students.txt', delimiter=',', skip_header=1, dtype=[('Student_Id', 'U40'), ('Assignment1_Marks', float), ('Assignment1_Time', 'U40'), ('Assignment2_Marks', float), ('Assignment2_Time', 'U40'), ('Assignment3_Marks', float), ('Assignment3_Time', 'U40'), ('Assignment4_Marks', float), ('Assignment4_Time', 'U40'), ('Assignment5_Marks', float), ('Assignment5_Time', 'U40'), ('Assignment6_Marks', float), ('Assignment6_Time', 'U40')])
print(Data)
```

### Find the mean value of assignment 1. Return this as a scalar value rounded to 2 decimal places

```python
MeanAssignment1 = round(np.mean(Data["Assignment1_Marks"]), 2)

#printing
print("Mean Value Of Assignment 1 : ", MeanAssignment1)
```

### Find the top student in the assignment 2. Return the marks rounded to 2 decimal places with student ID. 

```python
#finding the index of the top student in the array according to the marks in second assignment
TopStudentAssignment2 = np.argmax(Data["Assignment2_Marks"])

#obtaining the marks of the top student in assignemnt2
MarksOfTopStudent= round(Data['Assignment2_Marks'][TopStudentAssignment2], 2)

#obtaining the id of the top student 
IdOfTopStudent = Data['Student_Id'][TopStudentAssignment2]

#printing
print("Marks Of The Top Student In Assignment 2 :", MarksOfTopStudent)
print("Id Of The Top Student In Assignment 2 :",IdOfTopStudent)
```

### Find the Students ID who got 90 or above marks in assignment 1, assignment 3 and assignment 6. Return the students ID as a NumPy 1D array

```python
#finding the index
StudentHavingMarks = np.logical_and(np.logical_and(Data['Assignment1_Marks']>=90, Data['Assignment3_Marks']>=90), Data['Assignment6_Marks']>=90,)

#finding the ids
IdsOfStudents= Data['Student_Id'][StudentHavingMarks]

#printing
print("Students Who Got 90 Or Above Marks In Assignment 1, Assignment 3 and Assignment 6:",IdsOfStudents)
```

### Find the number of students who got 80 or below marks in assignment 4, and 90 or above marks in assignment 5. Return a scalar number. 

```python
#finding the index
MarksInAssignments = np.logical_and(Data['Assignment4_Marks']<=80, Data['Assignment5_Marks']>=90)

#find the total count
TotalStudents = np.sum(MarksInAssignments)

#printing
print("Number Of Students Who Got 80 Or Below Marks In Assignment 4 And 90 Or Above Marks In Assignment 5:", TotalStudents)
```

### Find the top student overall. Return his ID and marks rounded to 2 decimal places as NumPy 1D array. 

```python
#finding the total Marks of all assignments
TotalMarks = Data['Assignment1_Marks']+Data['Assignment2_Marks']+Data['Assignment3_Marks']+Data['Assignment4_Marks']+Data['Assignment5_Marks']+Data['Assignment6_Marks']

#finding the index of student with heighest total marks
IndexTopStudent= np.argmax(TotalMarks)

#finding the marks of top student
TotalMarksOfTopStudent = round(TotalMarks[IndexTopStudent], 2)

#finding the id of the top student
IdOfTopStudent= Data['Student_Id'][IndexTopStudent]

#printing
print("Id Of Top Student :",IdOfTopStudent)
print("Total Marks Of Top Student :",TotalMarksOfTopStudent)
```

### Each assignment has the equal weightage which is 2. Find the weighted average of each student’s assignments grade rounded to 1 decimal place. Return a 2D array with each student’s ID and average grade.  

```python
GradesOfAllAssignments = np.array([Data['Assignment1_Marks'], Data['Assignment2_Marks'], Data['Assignment3_Marks'], Data['Assignment4_Marks'], Data['Assignment5_Marks'], Data['Assignment6_Marks']]).T

Weights = np.array([2, 2, 2, 2, 2, 2])
#finding the weighted graes of all Assignments

WeightedGrades = np.multiply(GradesOfAllAssignments, weights)
#finding the weighted average of all assignments
WeightedAverage = np.round(np.sum(WeightedGrades, axis=1) / np.sum(Weights), 1)

# Create a 2D array with each student's ID and average grade
IdAndAverageGrade = np.column_stack((Data['Student_Id'], WeightedAverage))

#Printing 
print("Student ID\t\t\t\tWeighted Average Grade")
for value in IdAndAverageGrade:
    print(value[0],"\t\t",value[1])
   
```
